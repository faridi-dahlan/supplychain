```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#karena product description kosong, maka kita bisa drop kolom tersebut
df_supchain = df_supchain.drop(columns=['Product Description'])
```

```python
df_supchain
```

```python
df_supchain.columns
```

```python
#untuk memudahkan analisa, kita buat uppercase pada kolom bertipe data string
for column in df_supchain.columns:
    if df_supchain[column].dtype == 'object':
        df_supchain[column] = df_supchain[column].apply(lambda x : x.upper() if isinstance(x, str) else x)
```

```python
df_supchain
```

```python
df_supchain.info()
```

```python
#karena order zipcode ada yang kosong, kita akan ganti dengan 00000
df_supchain['Order Zipcode'] = df_supchain['Order Zipcode'].fillna(000000)
df_supchain.info()
```

```python
#dan dikarenakan ada yang kosong juga di customer zipcode, maka kita akan ganti juga ke 000000
# untuk customer lname tidak apa kosong, dikarenakan mungkin customer hanya punya first name saja (ex : ridho,rizal) bukan (ex : ridho rahmatullah)
df_supchain['Customer Zipcode'] = df_supchain['Customer Zipcode'].fillna(00000)
df_supchain.info()
```
